# Results Notebook without Selection

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from os import listdir
from os.path import isfile, join
from os import getcwd 
```

```python
path = getcwd()

new_plot_path = path + '/output/gridsearch/plots/'
longpath = path + '/output/gridsearch/'

if not os.path.exists(new_plot_path):
    print(f'creating path: {new_plot_path}')
    os.makedirs(new_plot_path)

```

```python
grid_options = []
for i in range(1,11):
    for j in range(1,11):
        grid_options.append((i,j))
print(grid_options)
```

## Data

```python
data = {}
all_files = {}
```

```python

print(f'Files from path: {longpath} \n')
for option in grid_options:
    print(f'Files in option {option[0]}-{option[1]}:')
    file_names = [f for f in listdir(f'{longpath}{option[0]}-{option[1]}') if isfile(join(f'{longpath}{option[0]}-{option[1]}', f))]
# file_names.remove('.DS_Store') #-- stupid Mac file

    print(file_names)
    print('\n')

    all_files[option] = file_names
    
```

### Load

```python
for option in grid_options:
    files = all_files[option]
    print(option, files)
    gossip = int(option[0])
    groom = int(option[1])

    for name in files:
        if 'avg_phenotypes_over_pop' in name:
            print(name)
            data[(gossip, groom)] = np.load(f'{longpath}/{gossip}-{groom}/{name}', allow_pickle=True)
    
```

```python

# print(data)
last_gossip = []
last_tolerance = []

for gossip, groom in data:
    last_gossip.append(data[(gossip, groom)][-1][0])
    last_tolerance.append(data[(gossip, groom)][-1][1])
```

```python
data_df = pd.Series(data).reset_index()   
data_df.columns = ['Gossip efficiency', 'Groom efficiency', 'Values'] 

data_df['last_gossip'] = last_gossip
data_df['last_tolerance'] = last_tolerance
data_df.head()
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(20, 8))

ax1, ax2 = axes

ax1 = sns.heatmap(data_df.pivot('Groom efficiency', 'Gossip efficiency', 'last_gossip'), vmin=0, vmax=1, annot=True, linewidths=.5, ax=ax1)
ax2 = sns.heatmap(data_df.pivot('Groom efficiency', 'Gossip efficiency', 'last_tolerance'), vmin=0, vmax=1, annot=True, linewidths=.5, ax=ax2)

ax1.title.set_text('Last gossip probability')
ax2.title.set_text('last tolerance probability')

fig.savefig(os.path.join(new_plot_path, "gridsearch.png"),
            dpi=300, 
            format='png', 
            bbox_inches='tight')
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

sns.heatmap(data_df.corr(),vmin=-1, vmax=1, annot=True, linewidths=.5)

fig.savefig(os.path.join(new_plot_path, "correlations.png"),
            dpi=300, 
            format='png', 
            bbox_inches='tight')
```
